```python
# Function to remove punctuation

def remove_punctuation(word):
    output = ""
    for char in word:
        if not char in string.punctuation:
            output += char
    return output

# Function to remove numbers from keys in dictionary

def remove_numbers(word):
    output = ""
    for char in word:
        if not char.isdigit():
            output += char
    return output

def remove_articles(word):
    list_of_articles = {'a': '',
                        'an':'',
                        'and':'',
                        'the':'',
                        'this':'',
                        'they':'',
                        'their':'',
                        'there':'',
                        'them':'',
                        'that':'',
                        'thats':'',
                        'are':'',
                        'as':'',
                        'its':'',
                        'to':'',
                        'is':'',
                        'it':'',
                        'go':'',
                        'was':'',
                        'or':'',
                        'but':'',
                        'so':'',
                        'if':'',
                        'i':'',
                        'for':'',
                        'on':'',
                        'all':'',
                        'in':'',
                        'of':'',
                        'at':'',
                        'can':'',
                        'be':'',
                        'have':'',
                        'had':'',
                        'because':'',
                        'said':'',
                        'here':'',
                        'from':'',
                        'like':'',
                        'with':'',
                        'you':'',
                        'me':'',
                        'we':'',
                        'or':'',
                        'my':''}
    other = []
    for char in word.split():
        if char not in list_of_articles:
            other.append(char)
    return ' '.join(other)
```

```python
# Chinese Restaurants
df_chi = pd.read_csv('chinese.csv')
```

```python
# Chinese Restaurants

my_dict = {}
my_dict = col.defaultdict(lambda: 0, my_dict)

#for i in range(len(df_chi.index)):
for i in range(20):
    
    tempString = df_chi.loc[i, 'text']
    
    tempList = tempString.split()
    
    for j in range(len(tempList)):
        tempWord = remove_numbers(remove_punctuation(remove_articles(tempList[j].lower())))
        if (any(c.isalpha() for c in tempWord)):
            my_dict[tempWord] += 1

pd.Series(my_dict).sort_values()
```
